```python
# LRU Page Replacement

length = 4
reference_string = [1, 2, 1, 4, 5, 6, 3, 4, 6, 3, 7, 3, 1, 5, 3, 7, 3, 4, 2, 4, 1, 4, 5, 1]
frame, string, fault, is_page_fault = [], [], 0, 'No'

print("\nString|Frame →\t", end='')

for i in range(length):
    print(i, end=' ')

print("Fault\n   ↓\n")

for i in reference_string:
    if i not in frame:
        if len(frame) < length:
            frame.append(i)
            string.append(len(frame)-1)
        else:
            ind = string.pop(0)
            frame[ind] = i
            string.append(ind)
        is_page_fault = 'Yes'
        fault += 1
    else:
        string.append(string.pop(string.index(frame.index(i))))
        is_page_fault = 'No'
        
    print("   %d\t\t"%i,end='')

    for x in frame:
        print(x, end=' ')
        
    for x in range(length-len(frame)):
        print(' ', end=' ')

    print(" %s" % is_page_fault)

print("\nTotal Page Faults (LRU): %d" % (fault))
```

```python
# Optimal Page Replacement

length = 4
reference_string = [1, 2, 1, 4, 5, 6, 3, 4, 6, 3, 7, 3, 1, 5, 3, 7, 3, 4, 2, 4, 1, 4, 5, 1]
frame, fault, is_page_fault = [], 0, 'No'

print("\nString|Frame →\t",end='')

for i in range(length):
    print(i,end=' ')

print("Fault\n   ↓\n")

occur = [None for i in range(length)]

for i in range(len(reference_string)):
    if reference_string[i] not in frame:
        if len(frame) < length:
            frame.append(reference_string[i])
        else:
            for x in range(len(frame)):
                if frame[x] not in reference_string[i+1:]:
                    frame[x] = reference_string[i]
                    break
                else:
                    occur[x] = reference_string[i+1:].index(frame[x])
            else:
                frame[occur.index(max(occur))] = reference_string[i]
        fault += 1
        is_page_fault = 'Yes'
    else:
        is_page_fault = 'No'
        
    print("   %d\t\t"%reference_string[i],end='')

    for x in frame:
        print(x, end=' ')
        
    for x in range(length-len(frame)):
        print(' ', end=' ')

    print(" %s" % is_page_fault)

print("\nTotal Page Faults (Optimal): %d" % (fault))
```
