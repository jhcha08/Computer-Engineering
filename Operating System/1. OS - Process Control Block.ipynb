```python
class PCB:
    def __init__(self):
        self.result = None
        
    def name(self, name):
        self.name = name

    def PID(self, PID):
        self.PID = PID
        
    def status(self, status):
        self.status = status
```

```python
PCB1 = PCB()
PCB1.name('PCB1')
PCB1.PID('PID001')
PCB1.status('none')

PCB2 = PCB()
PCB2.name('PCB2')
PCB2.PID('PID002')
PCB2.status('none')

PCB3 = PCB()
PCB3.name('PCB3')
PCB3.PID('PID003')
PCB3.status('none')

PCB4 = PCB()
PCB4.name('PCB4')
PCB4.PID('PID004')
PCB4.status('none')

PCB5 = PCB()
PCB5.name('PCB5')
PCB5.PID('PID005')
PCB5.status('none')

PCB6 = PCB()
PCB6.name('PCB6')
PCB6.PID('PID006')
PCB6.status('none')

PCB7 = PCB()
PCB7.name('PCB7')
PCB7.PID('PID007')
PCB7.status('none')
```

```python
jobQueue = [PCB1, PCB2, PCB3, PCB4, PCB5, PCB6, PCB7]
temp_jobQueue = [PCB1, PCB2, PCB3, PCB4, PCB5, PCB6, PCB7]
running, ready, wait = [], [], []

runProcess = jobQueue[0]
running.append(runProcess)
runProcess.status = 'Process Running'

temp_jobQueue.pop(0)

for i in range(1,4):
    ready.append(jobQueue[i])
    temp_jobQueue.pop(0)

print('#%s - %s %s' % (runProcess.name, runProcess.PID, runProcess.status))
print('Running : %s' % runProcess.name)
print('Ready : %s %s %s' % (ready[0].name, ready[1].name, ready[2].name))
print('Wait : %s (HDD I/O Queue)' % (wait[0].name)) if len(wait) != 0 else print('Wait : ')
print('Job Queue : ', end= ' ')
for i in range(len(jobQueue)):
    print(jobQueue[i].name, end= ' ')


while len(jobQueue) != 0:
    
    ###################

    runProcess.status = 'HDD I/O Event'
    wait.append(runProcess)
    running.clear()

    print('\n')
    print('#%s - %s %s' % (runProcess.name, runProcess.PID, runProcess.status))
    print('Running : %s' % runProcess.name) if len(running) != 0 else print('Running : ')
    print('Ready : ', end= ' ')
    for i in range(len(ready)):
        print(ready[i].name, end= ' ')
    print('\nWait : %s (HDD I/O Queue)' % (wait[0].name)) if len(wait) != 0 else print('\nWait : ')
    print('Job Queue : ', end= ' ')
    for i in range(len(jobQueue)):
        print(jobQueue[i].name, end= ' ')

    ###################

    runProcess = ready.pop(0) if len(ready) != 0 else wait.pop(0)
    running.append(runProcess)
    runProcess.status = 'Process Running'
    if len(temp_jobQueue) != 0:
        ready.append(temp_jobQueue.pop(0))

    print('\n')
    print('#%s - %s %s' % (runProcess.name, runProcess.PID, runProcess.status))
    print('Running : %s' % runProcess.name) if len(running) != 0 else print('Running : ')
    print('Ready : ', end= ' ')
    for i in range(len(ready)):
        print(ready[i].name, end= ' ')
    print('\nWait : %s (HDD I/O Queue)' % (wait[0].name)) if len(wait) != 0 else print('\nWait : ')
    print('Job Queue : ', end= ' ')
    for i in range(len(jobQueue)):
        print(jobQueue[i].name, end= ' ')

    ###################
    
    if len(wait) != 0 and len(jobQueue) != 1:
        waitProcess = wait[0]
        waitProcess.status = 'HDD I/O finshed'

        print('\n')
        print('#%s - %s %s' % (waitProcess.name, waitProcess.PID, waitProcess.status))
        print('Running : %s' % runProcess.name) if len(running) != 0 else print('Running : ')
        print('Ready : ', end= ' ')
        for i in range(len(ready)):
            print(ready[i].name, end= ' ')
        print('\nWait : %s (HDD I/O Queue)' % (wait[0].name)) if len(wait) != 0 else print('\nWait : ')
        print('Job Queue : ', end= ' ')
        for i in range(len(jobQueue)):
            print(jobQueue[i].name, end= ' ')

    ###################

    runProcess.status = 'Process Terminated'
    running.pop(0)
    jobQueue.remove(runProcess)

    print('\n')
    print('#%s - %s %s' % (runProcess.name, runProcess.PID, runProcess.status))
    print('Running : %s' % runProcess.name) if len(running) != 0 else print('Running : ')
    print('Ready : ', end= ' ')
    for i in range(len(ready)):
        print(ready[i].name, end= ' ')
    print('\nWait : %s (HDD I/O Queue)' % (wait[0].name)) if len(wait) != 0 else print('\nWait : ')
    print('Job Queue : ', end= ' ')
    for i in range(len(jobQueue)):
        print(jobQueue[i].name, end= ' ')

    ###################
    
    if len(jobQueue) != 0:
        runProcess = ready.pop(0) if len(ready) != 0 else wait.pop(0)
        running.append(runProcess)
        runProcess.status = 'Process Running'
        if len(wait) != 0:
            ready.append(wait.pop(0))
    else:
        break

    print('\n')
    print('#%s - %s %s' % (runProcess.name, runProcess.PID, runProcess.status))
    print('Running : %s' % runProcess.name) if len(running) != 0 else print('Running : ')
    print('Ready : ', end= ' ')
    for i in range(len(ready)):
        print(ready[i].name, end= ' ')
    print('\nWait : %s (HDD I/O Queue)' % (wait[0].name)) if len(wait) != 0 else print('\nWait : ')
    print('Job Queue : ', end= ' ')
    for i in range(len(jobQueue)):
        print(jobQueue[i].name, end= ' ')
```
