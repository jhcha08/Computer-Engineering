```python
# 8958

a = int(input())

num_list = []

list_ = [None]*a

for k in range(len(list_)):
    list_[k] = input().split()[0]
    
for index in list_:
    num = 0
    list_o = []
    for i in range(len(index)):
        if index[i] == 'O':
            list_o.append(index[i])
            num += len(list_o)
        elif index[i] == 'X':
            list_o = []
    num_list.append(num)

for n in num_list:
    print(n)
```

```python
# 4344

a = int(input())
lista = [None]*a
list_score = []

for k in range(len(lista)):
    lista[k] = input().split()
    list_ = [int(i) for i in lista[k]]
    
    list_h = []
    mean = sum(list_[1:])/list_[0]
    
    for s in list_[1:]:
        if s > mean:
            list_h.append(s)
        score = (len(list_h)/len(list_[1:]))*100
        
    list_score.append(score)
    
for sc in list_score:
    print('%0.3f%%' % sc)
```
